```python
#특정영화와 비슷한 영화추천
import pandas as pd
import numpy as np
```

```python
rating_data = pd.read_csv('movie-master/ratings.csv')
movie_data = pd.read_csv('movie-master/movies.csv')
```

```python
rating_data.head(5)
```

```python
movie_data.head(5)
```

```python
rating_data.drop('timestamp', axis = 1, inplace = True) #불필요한 컬럼제거(drop)
rating_data.head(5)
```

```python
movie_data.drop('genres',axis=1,inplace = True)
movie_data.head()
```

```python
user_movie_data= pd.merge(rating_data, movie_data,on='movieId')
user_movie_data.head()  # movieID를 기준으로 pd.merge
```

```python
# pivot table 만들기
# data : 영화 평점 rating
# index : 영화 title
# columns : userID
```

```python
user_movie_data.shape
```

```python
user_movie_rating = user_movie_data.pivot_table('rating',index='userId',columns='title').fillna(0)
```
